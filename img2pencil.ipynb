```python
import cv2
from matplotlib import pyplot as plt
from IPython.display import display, Image
```

```python
image = cv2.imread("datasets/dog.jpg")
```

```python
def im_show(image):
    display(Image(data=cv2.imencode('.jpg', image)[1].tobytes()))
```

```python
im_show(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
im_show(gray)
```

```python
# Method -1 : Invert image using operation
inv = 255-gray
im_show(inv)
```

```python
inv = cv2.bitwise_not(gray) # invert the image meaning 255-image
im_show(inv)
```

```python
# smoothing
img_smooth = cv2.GaussianBlur(inv, (5, 5), sigmaX=0, sigmaY=0)
# (5, 5) is Gaussian width and hight, must be positive and odd values
im_show(img_smooth)
```

```python
# pencil sketch  = gray / inverted pixel
res = cv2.divide(gray, 255 - img_smooth, scale=255)

# res = gray/(255 - img_smooth)

im_show(res)
```
